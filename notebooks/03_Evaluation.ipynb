```python
import pandas as pd
```

```python
qwen = pd.read_csv("data/qwen_results.csv")
chatgpt = pd.read_csv("data/chatgpt_results.csv")
```

```python
qwen = qwen.rename(columns={"response": "qwen_response"})
chatgpt = chatgpt.rename(columns={"response": "chatgpt_response"})
```

```python
comparison = qwen.merge(
    chatgpt[["id", "chatgpt_response"]],
    on="id"
)
```

```python
comparison.head()
```

```python
comparison["qwen_score"] = ""
comparison["chatgpt_score"] = ""
comparison["winner"] = ""
comparison["remarks"] = ""

comparison.to_csv("data/comparison_results.csv", index=False)
```

```python
comparison.to_csv(
    "data/comparison_results.csv",
    index=False
)

print("Comparison file created successfully!")
```
